```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = "resources/2019data.csv"
data2020 = "resources/NYPD_Arrest_Data__Year_to_Date_.csv"
nypd_data = pd.read_csv(data, parse_dates=["ARREST_DATE"])
nypd_data2020 = pd.read_csv(data2020, parse_dates=["ARREST_DATE"])
```

```python
april_data_19 = nypd_data.loc[(nypd_data["ARREST_DATE"] > "04/01/2019") & (nypd_data["ARREST_DATE"] < "05/01/2019"), :]
april_data_19.head()
```

```python
group_ofns_19 = april_data_19.groupby("OFNS_DESC")
count_ofns_19 = group_ofns_19["OFNS_DESC"].count().sort_values(ascending=False).head(15)
count_ofns_19
```

```python
plt.figure(figsize=(5,10))

ofns_chart_19 = count_ofns_19.sort_values().plot(kind="barh", alpha=0.5, fontsize=10)
plt.title("Top 15 Crimes April 2019", fontsize=10)
plt.ylabel("Offense Type", fontsize=10)
plt.xlabel("Count", fontsize=10)

plt.show()
```

```python
april_data_20 = nypd_data2020.loc[(nypd_data2020["ARREST_DATE"] > "04/01/2020") & (nypd_data2020["ARREST_DATE"] < "05/01/2020"), :]
april_data_20.head()
```

```python
group_ofns_20 = april_data_20.groupby("OFNS_DESC")
count_ofns_20 = group_ofns_20["OFNS_DESC"].count().sort_values(ascending=False).head(15)
count_ofns_20
```

```python
plt.figure(figsize=(5,10))

ofns_chart_20 = count_ofns_20.sort_values().plot(kind="barh", alpha=0.5, fontsize=10)
plt.title("Top 15 Crimes April 2020", fontsize=10)
plt.ylabel("Offense Type", fontsize=10)
plt.xlabel("Count", fontsize=10)

plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))

plt.subplot(1, 2, 1)

ofns_chart_20 = count_ofns_20.sort_values().plot(kind="barh", alpha=0.5, fontsize=10)
plt.title("Top 15 Crimes April 2020", fontsize=10)
plt.ylabel("Offense Type", fontsize=10)
plt.xlabel("Count", fontsize=10)

plt.subplot(1, 2, 2)
ofns_chart_19 = count_ofns_19.sort_values().plot(kind="barh", alpha=0.5, fontsize=10)
plt.title("Top 15 Crimes April 2019", fontsize=10)
plt.ylabel("Offense Type", fontsize=10)
plt.xlabel("Count", fontsize=10)

fig.tight_layout()
plt.show()
```
