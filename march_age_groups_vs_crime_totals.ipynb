```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = "resources/NYPD_Arrest_Data__Year_to_Date_.csv"
nypd_data = pd.read_csv(data)
nypd_data.shape
```

```python
nypd_data.head()
```

```python
nypd_data_1 = nypd_data.loc[:, ["ARREST_DATE", "PD_CD", "PD_DESC", "OFNS_DESC", "AGE_GROUP"]]
nypd_data_1.head(10)
```

```python
march_data = nypd_data_1.loc[(nypd_data_1["ARREST_DATE"] > "03/01/2020") & (nypd_data_1["ARREST_DATE"] < "04/01/2020"), :]
march_data.head(10)
```

```python
march_age_18_24 = march_data.loc[(march_data["AGE_GROUP"] == "18-24"), :]
march_age_18_24.head(10)
```

```python
group_ofns = march_age_18_24.groupby("OFNS_DESC")
count_ofns = group_ofns["OFNS_DESC"].count()
count_ofns
```

```python
plt.figure(figsize=(40,10))

ofns_chart_18_24m = count_ofns.plot(kind="bar", alpha=0.5)
plt.title("Age: 18-24")
plt.xlabel("OFNS TYPE")



plt.show()
```

```python
march_age_25_44 = march_data.loc[(march_data["AGE_GROUP"] == "25-44"), :]
group_ofns_25_44 = march_age_25_44.groupby("OFNS_DESC")
count_ofns_25_44 = group_ofns["OFNS_DESC"].count()
count_ofns_25_44
```
